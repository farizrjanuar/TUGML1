# Ambil Data

```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Download latest version
path = kagglehub.dataset_download("anandshaw2001/taxi-trip-dataset")

print("Path to dataset files:", path)
```

```python
import os

# List files in the dataset directory
dataset_files = os.listdir(path)
print("Files in dataset:", dataset_files)
```

```python
df = pd.read_csv(os.path.join(path, "Taxi_Trip_Data.csv"))

# Tampilkan beberapa baris pertama
df.head()
```

Informasi bentuk data

```python
df.info()
```

Melihat deskripsi statistik

```python
df.describe()
```

# Visualisasi

```python
# Visualisasi 1: Distribusi Jumlah Penumpang per Vendor
plt.figure(figsize=(10, 6))
sns.countplot(x='VendorID', hue='passenger_count', data=df)
plt.title('Distribusi Jumlah Penumpang per Vendor')
plt.xlabel('Vendor ID')
plt.ylabel('Jumlah Perjalanan')
plt.show()
```

Informasi yang didapat adalah banyak penumpang yang di ambil vendor

```python
# Visualisasi 2:  Hubungan antara Jarak Tempuh dan Biaya Perjalanan
plt.figure(figsize=(10, 6))
sns.scatterplot(x='trip_distance', y='total_amount', data=df)
plt.title('Hubungan Jarak Tempuh dan Biaya Perjalanan')
plt.xlabel('Jarak Tempuh (mil)')
plt.ylabel('Biaya Total ($)')
plt.show()
```

Informasi yang didapat adalah biaya dari jarak yang ditempuh, untuk data yang jauh termasuk kedalam anomali

```python
# Visualisasi 3: Distribusi Biaya Perjalanan
plt.figure(figsize=(10, 6))
sns.histplot(df['total_amount'], bins=30, kde=True)
plt.title('Distribusi Biaya Perjalanan')
plt.xlabel('Biaya Total ($)')
plt.ylabel('Frekuensi')
plt.show()
```

Informasi yang didapat adalah frekuensi besar biaya perjalanan
